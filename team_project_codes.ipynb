# Random Forest

```python
import pandas as pd

train_df = pd.read_csv("./data/preprocessed_train.csv")
test_df = pd.read_csv("./data/preprocessed_test.csv")
```

```python
y_train = train_df["class"]
x_train = train_df.drop(["mushroom_id","class"], axis=1)
x_test = test_df.drop(["mushroom_id"], axis=1)
```

```python
test_df
```

```python
from sklearn.model_selection import cross_val_score, KFold

kf = KFold(n_splits=5, shuffle=True, random_state=1234)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=10,
    criterion='entropy',
    max_depth=5,
    random_state=1234

)
```

```python
scores = cross_val_score(model, x_train, y_train, cv=kf)
```

```python
scores, scores.mean()
```

```python
model.fit(x_train, y_train)
```

```python
from matplotlib import pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(15,5))
_= plot_tree(
    model.estimators_[3],
    feature_names=x_train.columns,
    filled=True,
    rounded=True,
    fontsize=6
)
```

```python
from sklearn.model_selection import StratifiedKFold

stratified_kf= StratifiedKFold(n_splits=5, shuffle=True, random_state=1234)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [2,4,6],
    'max_depth': [3,4,5],
    'max_features': [6,7,8]

}

grid_search = GridSearchCV(
    estimator = model,
    param_grid = param_grid,
    cv=stratified_kf

)

grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_, grid_search.best_score_
```

```python
model = RandomForestClassifier(
    n_estimators=2,
    criterion='entropy',
    max_depth=5,
    max_features=8,
    random_state=1234

)
```

```python
scores = cross_val_score(model, x_train, y_train, cv=stratified_kf)
```

```python
scores.mean()
```

```python
model.fit(x_train,y_train)
```

```python
y_test = model.predict(x_test)
```

```python
test_df["class"] = y_test
```

```python
test_df[["mushroom_id","class"]].to_csv("./data/submission_rf.csv", index=False)
```

```python
import pandas as pd

train_df = pd.read_csv("./data/preprocessed_train.csv")
test_df = pd.read_csv("./data/preprocessed_test.csv")

y_train = train_df["class"]
x_train = train_df.drop(["mushroom_id","class"], axis=1)
x_test = test_df.drop(["mushroom_id"], axis=1)
```

```python
from sklearn.model_selection import StratifiedKFold

stratified_kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=1234)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [10,30],
    'learning_rate': [0.01,0.001],
    'num_leaves': [10,20,30],
    'subsample' : [0.8,1.0]

}
```

```python
import lightgbm as lgb

model = lgb.LGBMClassifier()
```

```python
_=grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=stratified_kf,
    scoring='accuracy'


)

_=grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
model =lgb.LGBMClassifier(**grid_search.best_params_)
```

```python
model.fit(x_train,y_train)
```

```python
from lightgbm import plot_importance

plot_importance(model)
```
